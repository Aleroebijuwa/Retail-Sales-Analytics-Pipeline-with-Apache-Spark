# Spark Setup Test

Verifies the local Apache Spark 3.5.0 / PySpark environment: creates a `SparkSession`, builds a small DataFrame, and counts its rows.

Run with the **Spark (PySpark 3.5)** kernel.

```python
# Test your PySpark installation with this notebook cell
from pyspark.sql import SparkSession

# Create SparkSession
spark = SparkSession.builder \
    .appName("SparkSetupTest") \
    .master("local[*]") \
    .getOrCreate()

# Create a simple test DataFrame
data = [(1, "Alice", 100), (2, "Bob", 200), (3, "Charlie", 150)]
columns = ["id", "name", "amount"]
df = spark.createDataFrame(data, columns)

# Display DataFrame
df.show()

# Print Spark version and row count
print(f"Spark Version: {spark.version}")
print(f"Row Count: {df.count()}")

# Stop the session
spark.stop()
```
